```python
# importing dependencies
import pandas as pd
import numpy as np
```

```python
# merge

emp = {
    "emp_id" : [101, 102, 103, 104, 105, 106],
    "emp_name" : ["Sam", "Mike", "Alex", "John", "Priya", "Rex"],
    "salary" : [25000, 45000, 35000, 45000, np.nan, np.nan],
    "dept_id" : [1, 2, 3, np.nan, 2, np.nan]
}

# dept table
dept = {"department_id" : [1, 2, 3, 4],
 "dept_nane" : ["IT", "HR", "Marketing", "Operation"],
 "region" : ["India", "UK", "US", "Canada"]
 }

df_emp = pd.DataFrame(emp)
df_dept = pd.DataFrame(dept)
```

```python
df_dept
```

```python
df_emp
```

 merge - same as join we have in SQL  
it combined two or more dataframe together  

```python
pd.merge(df_emp, df_dept, how='inner', left_on='dept_id', right_on='department_id')
```

```python
pd.merge(df_emp, df_dept, how='left', left_on='dept_id', right_on='department_id')
```

```python
pd.merge(df_emp, df_dept, how='right')
```

```python
pd.merge(df_emp, df_dept, how='outer')
```

**CONCAT**

```python
df1 = pd.DataFrame({
    'emp_id': [1, 2, 3],
    'emp_name': ['Alice', 'Bob', 'Charlie'],
    'salary': [50000, 60000, 55000]
})

df2 = pd.DataFrame({
    'emp_id': [4, 5],
    'emp_name': ['David', 'Eve'],
    'salary': [70000, 45000]
})
```

```python
pd.concat(([df_dept, df_emp]), axis=0)
```

```python
df_dept
```

```python
pd.concat(([df1, df2]), axis=1)
```

```python
import pandas as pd

data = {
    'city': ['New York', 'Los Angeles', 'New York', 'Los Angeles', 'New York',],
    'product': ['A', 'A', 'B', 'B', 'A'],
    'total': [100, 200, 150, 300, 750]
}
df = pd.DataFrame(data)
```

```python
df
```

GROUP BY 

```python
df.groupby(by='city')['total'].count()
```

```python
df.groupby(by='city')['total'].sum()
```

```python
df.groupby(by='city').agg({'total' : ['sum', 'count', 'mean']})
```

```python
df.groupby(by=['city','product']).agg({'total' : ['sum', 'count', 'mean']})
```

```python
df.groupby(['city','product'])['total'].count()
```

Q.1 Find out each city and total no of orders

```python
df.groupby('city').agg({
                        'total' : ['count', 'sum'], 
                       'quantity' : 'sum'
                       })
```

```python
df.groupby(['branch', 'gender']).agg({'total' : ['count', 'sum']})
```

```python
df.groupby('rating')['invoice_id'].count()
```

```python
df.groupby('rating').agg({'invoice_id' : 'count'})
```
