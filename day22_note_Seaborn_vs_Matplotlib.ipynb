**What is Seaborn?**  
Seaborn is a Python library built on Matplotlib for making attractive, easy-to-read statistical plots. It simplifies complex visualizations and works well with Pandas DataFrames.


### **Seaborn vs. Matplotlib**

| Feature                  | Seaborn                          | Matplotlib                     |
|--------------------------|----------------------------------|--------------------------------|
| Ease of Use              | High (less code required)        | Moderate (more setup required) |
| Default Style            | Beautiful and consistent         | Basic                          |
| Support for Statistics   | Built-in                         | Requires additional work       |
| Integration with Pandas  | Excellent                        | Limited                        |

- Scatter Plots (Relationships)  
- Histograms (Distribution)  
- Bar Plots (Categorical Data)  
- Heatmaps (Correlations)  

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# datasets
iris = pd.read_csv('iris.csv', encoding_errors='ignore')
car_crashes = pd.read_csv('car_crashes.csv', encoding_errors='ignore')
flights = pd.read_csv('flights.csv', encoding_errors='ignore')
taxis = pd.read_csv('taxis.csv', encoding_errors='ignore')
tips = pd.read_csv('tips.csv', encoding_errors='ignore')

# printing shape
print(f"Iris Dataset Shape {iris.shape}")
print(f"car Dataset Shape {car_crashes.shape}")
print(f"flight Dataset Shape {flights.shape}")
print(f"taxis Dataset Shape {taxis.shape}")
print(f"tips datasets shape {tips.shape}")
```

**Scatter Plot**

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time')
```

Explanation:
data: The dataset used (e.g., tips).
x and y: Columns representing x and y axes.
hue: Groups data by color (e.g., smoker).
style: Differentiates data points by markers (e.g., time).

**Box Plot**

Box Plot using sns.boxplot()
Task: Compare total bill distribution across different days.

**Heatmap**

Heatmap using sns.heatmap()
Task: Visualize the correlation matrix of the tips dataset.

```python
tips_df = tips[['total_bill', 'tip', 'size']]
tips_df.head()
```

```python
sns.heatmap(data=tips_df.corr(), annot=True)
```

**Bar Plot**

Bar Plot using sns.barplot()
Task: Compare average total bills across different days.

```python
sns.barplot(data=tips, x='day', y='total_bill', ci=None)
plt.show()
```

Explanation:
ci=None: Disables confidence intervals.

**Join Plot**

Joint Plot using sns.jointplot()
Task: Create a joint plot between total bill and tip.

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
plt.show()
```

Explanation:
kind='reg': Adds a regression line.

**Swarm Plot**

Swarm Plot using sns.swarmplot()
Task: Visualize the spread of total bills by day.

```python
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', palette='husl')
plt.show()
```

Explanation:
palette: Controls color schemes.
